```
import pandas as pd
```

```
df = pd.read_csv('/avocado.csv')
```

```
df['Date']=pd.to_datetime(df['Date'])
```

```
df.set_index('Date',inplace=True)
```

```
df.sort_index(ascending=True,inplace=True)
```

```
cf = df.copy()[df['type']=='conventional']
```

```
region=list(cf['region'].unique())
```

```
albany_df = cf[cf['region']=='Albany']
```

```
albany_df['AveragePrice'].rolling(25).mean().dropna().plot()
```

```
Graph_df = pd.DataFrame
for r in region:
  r_df = cf.copy()[cf['region']==r]
  r_df[f'{r}_25ma'] = r_df['AveragePrice'].rolling(25).mean()
  
  if Graph_df.empty:
    Graph_df = r_df[[f'{r}_25ma']] # double big brackets so that it doesn't identify it as a series since we can't use join with a series.
  else:
    Graph_df = Graph_df.join(r_df[f'{r}_25ma'])
```

```
Graph_df.dropna().plot(figsize=(8,5),legend=False)
```
