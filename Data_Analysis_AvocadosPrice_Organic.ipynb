```
regions=list(df['region'].unique())
```

```
regions
```

```
graph_df = pd.DataFrame
for r in regions:
  r_df = df.copy()[df['region']==r]
  r_df.set_index("Date", inplace=True)
  r_df.sort_index(inplace=True)
  r_df[f"{r}_25MA"] = r_df['AveragePrice'].rolling(25).mean() #here the 'f' in the r_df[f"{r}_25MA"] is a f string used as an replacement to .format in python.

  if graph_df.empty:
    graph_df = r_df[[f"{r}_25MA"]]#double brackets to prevent series since single brackets make it a series and we wont be able to use join later on.
  else:
    graph_df = graph_df.join(r_df[f"{r}_25MA"])

graph_df.tail()
```

```
graph_df.plot(figsize=(10,7),legend=False)
```
